```python
import requests
```

```python
a= requests.get("https://quotes.toscrape.com/")
```

```python
with open ("Htmls/quote.html","w",encoding="utf-8") as f:
    f.write(a.text)
```

```python
from bs4 import BeautifulSoup
```

```python
with open ("Htmls/quote.html",encoding="utf-8") as f:
    content=f.read()
```

```python
soup=BeautifulSoup(content,"html.parser")
```

```python
span=soup.find_all("span")
```

```python
items=[]
for sp in span:
    quote=sp.find_all(class_="text")
    Author=sp.select_one("small").text
    items.append([quote,Author])
```

```python
import pandas as pd
```

```python
df=pd.DataFrame(items,columns=['Quote','Author'])
```

```python
import matplotlib.pyplot as plt
```

```python
# sachin tendulkar's runs
years = [1990, 1992, 1994, 1996, 1998, 2000, 2003, 2005, 2007, 2010]
runs =  [500, 700, 1100, 1500, 1800, 1200, 1700, 1300, 900, 1500]

plt.bar(years, runs)
plt.xlabel("Years")
plt.ylabel("Runs")
plt.title("Sachin tendulakars yearly runs")
```

```python
# plot 3 bars side by side 
sachin =  [500, 700, 1100, 1500, 1800, 1200, 1700, 1300, 900, 1500]
kohli = [0, 0, 500, 800, 1100, 1300, 1500, 1800, 1900, 2100]
sehwag = [0, 300, 800, 1200, 1500, 1700, 1600, 1400, 1000, 0]
years = [1990, 1992, 1994, 1996, 1998, 2000, 2003, 2005, 2007, 2010]

import numpy as np
x=np.arange(len(years))
width=0.25
plt.bar(x-width,sachin,width=width,label="Sachin runs")
plt.bar(x,kohli,width=width,label="Virat runs")
plt.bar(x+width,sehwag,width=width,label="Sehwag runs")
plt.xlabel="Years"
plt.ylabel="Runs"
plt.title="Run comparision"
plt.xticks(x,years) # show years instant of 0,1,2,3.. 
plt.legend()
plt.tight_layout()
```

```python
# Add text label in bar
players = ["Sachin", "Sehwag", "Kohli"]
runs = [1500, 1200, 1800]

plt.bar(players, runs, color="skyblue")

# Add labels on top of bars
for i in range(len(players)):
    plt.text(i, runs[i] + 50, f"{runs[i]}", ha='center',color="black")

plt.ylabel('runs')
plt.title("Runs Scored by Players")
plt.tight_layout()
plt.show()
```

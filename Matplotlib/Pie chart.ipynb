```python
import matplotlib.pyplot as plt
```

```python
# simle pie chart 
plt.style.use('ggplot')
labels = ["Sachin", "Sehwag", "Kohli", "Yuvraj"]
runs = [18000, 8000, 12000, 9500]
plt.pie(runs,labels=labels)
plt.title("Carrer runs")
```

```python
# for giving our own color
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99']
plt.pie(runs,labels=labels,colors=colors)
```

```python
# use of wedgeprops 
# it use for giving some style to our pie chart
plt.pie(
     runs,
     labels=labels,
     colors=colors,
     wedgeprops={'edgecolor':'#127809' ,'linestyle':':','linewidth':2}
)
```

```python
# use of explode
# explode is use of highlight perticular section from pie chart
explode=[0.1,0,0,0] # highlight only sachin's slice
plt.pie(runs,
        labels=labels,
        colors=colors,
        explode=explode,
        shadow=True
       )
```

```python
# for rotate pie chart and use of autopct
explode=[0.1,0,0,0] # highlight only sachin's slice
plt.pie(runs,
        labels=labels,
        colors=colors,
        explode=explode,
        shadow=True,
        startangle=140,
        autopct="%1.f%%" #autopct is use for giving percentage value to pie chart 
       )
```

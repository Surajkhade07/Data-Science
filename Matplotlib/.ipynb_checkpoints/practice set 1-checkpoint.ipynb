```python
import matplotlib.pyplot as plt
```

```python
plt.grid(True) # for giving grid to our graph
kohli = [0, 0, 500, 800, 1100, 1300, 1500, 1800, 1900, 2100]
sehwag = [0, 1000, 800, 1200, 1500, 1700, 1600, 1400, 1000, 0]
rohit =[500,1200,900,1000,300,1800,1700,1500,1800,2000]
years = [1990, 1992, 1994, 1996, 1998, 2000, 2003, 2005, 2007, 2010]
plt.xlabel("Years")
plt.ylabel("Runs")
plt.plot(years,kohli,"ro--",label="virat runs")
plt.plot(years,rohit,"bv:",label="rohit runs")
plt.plot(years,sehwag,"g^:",label="sehwag runs")
plt.title("Runs comparison")
plt.legend()
```

```python
# probelm of pie chart
import matplotlib.pyplot as plt
mobileos=["Android","IOS","Others"]
shares=[75.85,25.26,10]
colors=['#ffa233','#33ff83','#bb33ff']
plt.pie(shares,labels=mobileos,colors=colors,autopct="%1.1f%%")
```

```python
# same task using harizontal bar chart
mobileos=["Android","IOS","Others"]
shares=[75.85,25.26,10]
plt.barh(mobileos,shares,color='skyblue')
```

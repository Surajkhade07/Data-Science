```python
# sachin tendulkar's runs
years = [1990, 1992, 1994, 1996, 1998, 2000, 2003, 2005, 2007, 2010]
runs =  [500, 700, 1100, 1500, 1800, 1200, 1700, 1300, 900, 1500]

plt.plot(years, runs)
```

```python
# Giving labels to x and y axis
years = [1990, 1992, 1994, 1996, 1998, 2000, 2003, 2005, 2007, 2010]
runs =  [500, 700, 1100, 1500, 1800, 1200, 1700, 1300, 900, 1500]

plt.plot(years, runs)
plt.xlabel("Years")
plt.ylabel("Runs")
plt.title("Sachin tendulkar yearly runs")
```

```python
# Multipel lines in plot
plt.grid(True) # for giving grid to our graph
kohli = [0, 0, 500, 800, 1100, 1300, 1500, 1800, 1900, 2100]
sehwag = [0, 300, 800, 1200, 1500, 1700, 1600, 1400, 1000, 0]
years = [1990, 1992, 1994, 1996, 1998, 2000, 2003, 2005, 2007, 2010]
plt.plot(years,kohli,label="Virat Runs")
plt.plot(years,sehwag,label="sehwag Runs")
plt.xlabel("Years")
plt.ylabel("Run score")
plt.title("Performance plot")
plt.legend()
```

```python
# Giving Format string to plot
# for giving format string :[color][markar][line]
plt.plot(years,kohli,"rv--",label="Virat Runs") # red circles with dash line  
plt.plot(years,sehwag,"bv:",label="sehwag Runs") # green triangle with doted line
```

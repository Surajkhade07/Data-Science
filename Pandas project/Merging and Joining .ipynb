```python
import pandas as pd
```

```python
employees=pd.DataFrame({
 "empID":[1,2,3],
 "Name":["Suraj","Swami","Kunal"],
 "DepID":[10,20,30]
})
```

```python
employees
```

```python
department=pd.DataFrame({
  "DepID":[10,20,40],
  "Depname":["HR","IT","devops"]
})
```

```python
department
```

```python
# merging of two columns with there common column
pd.merge(employees,department, on="DepID") # Inner join 
```

```python
pd.merge(employees,department, on="DepID",how="left") # left join  
```

```python
pd.merge(employees,department, on="DepID",how="right") # right join 
```

```python
pd.merge(employees,department, on="DepID",how="outer") # outer join 
```

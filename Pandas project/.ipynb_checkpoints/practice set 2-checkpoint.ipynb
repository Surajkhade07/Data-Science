```python
import pandas as pd
```

```python
df=pd.read_json("data2.json")
```

```python
df
```

```python
# 1st problem find avg marks of each subject per class 
df.groupby("Class")[["Maths","Science","English"]].mean()
```

```python
#2nd problem which class has highest avg total marks
# add new column of total 
df["Total"]=df[["Maths","Science","English"]].sum(axis=1)
```

```python
df
```

```python
# get the avg of total marks of per class
avg_total_class=df.groupby("Class")["Total"].mean()
```

```python
avg_total_class
```

```python
#get class with highest avg total marks
high_avg_total=avg_total_class.idxmax()
```

```python
print(f"class with highest avg total marks is:{high_avg_total}")
```
